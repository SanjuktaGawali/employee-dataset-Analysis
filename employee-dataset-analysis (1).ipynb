```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Summary**- The analysis of the employee dataset has yielded valuable insights into the composition and dynamics of the workforce. By examining various dimensions such as education, joining year, geographical distribution, payment tier, age, gender, experience, and leave-taking behavior, a comprehensive view of the organization's human resources has been achieved.

```python
df1 = pd.read_csv("/kaggle/input/employee-dataset")
df1
```

```python
df1.info()
```

**What is the distribution of educational qualifications among employees?**

```python
# Countplot for Education distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='Education', data=df1)
plt.title('Distribution of Educational Qualifications Among Employees')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=360)
plt.show()
```

**How does the length of service (Joining Year) vary across different cities?**

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='City', y='JoiningYear', data=df1, palette='viridis')
plt.title('Variation of Length of Service Across Different Cities')
plt.xlabel('City')
plt.ylabel('Joining Year')
plt.xticks(rotation=360)
plt.show()
```

**Is there a correlation between Payment Tier and Experience in Current Domain?**

```python
cross_tab = pd.crosstab(df1['PaymentTier'], df1['ExperienceInCurrentDomain'])
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, annot=True, cmap='Blues', fmt='d')
plt.title('Correlation Between Payment Tier and Experience in Current Domain')
plt.xlabel('Experience in Current Domain')
plt.ylabel('Payment Tier')
plt.show()
```

**What is the age distribution by gender within the workforce?**

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df1, x='Age', hue="Gender", palette='pastel')

plt.title('Age Distribution by gender Within the Workforce')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Are there any patterns in leave-taking behavior among employees?**

```python
leave_distribution = df1['LeaveOrNot'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(6, 6))
plt.pie(leave_distribution, labels=leave_distribution.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Leave-Taking Behavior Among Employees')
plt.show()
```

**Conclusion** - The analysis of the employee dataset revealed valuable insights into various aspects of the workforce. Key findings include a distribution of educational qualifications among employees, patterns in leave-taking behavior, and the exploration of factors such as joining year, city, payment tier, age, gender, and experience in the current domain. The visualizations, including bar charts and pie charts, provided a clear and concise representation of the data, aiding in the identification of trends and patterns.
